```python
# list comprehension with if condition
even_numbers = [x for x in range(1, 11) if x % 2 == 0]
print(even_numbers)
```

```python
# lambda function 
add = lambda x, y: x + y
result = add(5, 3)
print(result)
```

```python
from functools import reduce

# map() to double each element in a list
numbers = [1, 2, 3, 4, 5]
doubled_numbers = list(map(lambda x: x * 2, numbers))
print(doubled_numbers)

# filter() to get even numbers from a list
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)

# reduce() to calculate the sum of elements in a list
sum_of_numbers = reduce(lambda x, y: x + y, numbers)
print(sum_of_numbers)
```

```python
# Year-wise average score
print("year-wise avg marks")
year_avg_score = df.groupby('Year')['Marks_scored'].mean()
print(year_avg_score)
```

```python
# Semester-wise average score
print("semster-wise avg marks")
semester_avg_score = df.groupby('Semester')['Marks_scored'].mean()
print(semester_avg_score)
```

```python
def categorize_marks(Marks_scored):
    if Marks_scored >= 90:
        return 'Excellent'
    elif Marks_scored >= 80:
        return 'Very Good'
    elif Marks_scored >= 70:
        return 'Good'
    else:
        return 'Average'

df['Mark_Category'] = df['Marks_scored'].apply(categorize_marks)

# Display the updated data frame
print(df)
```

```python
# mean, median and standard deviations of semester wise marks.
semester_statistics = df.groupby('Semester')['Marks_scored'].agg(['mean', 'median', 'std'])
print(semester_statistics)
```
